```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
print("\n Dataset Shape:", df.shape)
print("\n First 5 Rows:\n", df.head())
print("\n Data Types:\n", df.dtypes)
print("\n Missing Values:\n", df.isnull().sum())
```

```python
#  Visualization 1: Survival Count
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.xlabel("Survived")
plt.ylabel("Passenger Count")
plt.show()
```

```python
# Visualization 2: Survival by Gender
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
#  Visualization 3: Survival by Passenger Class
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#  Visualization 4: Age Distribution by Survival
plt.figure(figsize=(8,5))
sns.kdeplot(df[df['Survived']==1]['Age'].dropna(), label='Survived', shade=True)
sns.kdeplot(df[df['Survived']==0]['Age'].dropna(), label='Did Not Survive', shade=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.legend()
plt.show()
```

```python
#  Observations
print("\n Insights:")
print("1. More females survived compared to males.")
print("2. First class passengers had higher survival rates.")
print("3. Children (Age < 15) had relatively better survival chances.")
print("4. Most missing values are in Age and Cabin columns.")
```
